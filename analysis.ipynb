## Question1 : 
How many employees spent above their given stipend? Extract the name of the
employee, the assigned stipend, and the actual amount spent.

```python
import psycopg2

import pandas as pd
```

```python
conn = psycopg2.connect(
    host = 'localhost'
    , user = 'princess'
    , password = 'Bismillahi@123'
    , database = 'bissan'
)
cursor = conn.cursor()
```

```python
query = '''
SELECT DISTINCT name , entitled_bonus AS stipend
FROM employees e JOIN departments d
ON e.department_id = d._id
'''
```

```python
cursor.execute(query)
all_employee_stipends = cursor.fetchall()
```

```python
employee_df =pd.DataFrame(all_employee_stipends,columns=['name','stipends'])
```

```python
employee_df = employee_df.drop_duplicates(subset='name') #froping duplicates
```

```python
travels_dataset = pd.read_csv('Cleaned_Travel_dataset.csv') 
```

```python
travels_dataset = travels_dataset.drop('Unnamed: 0',axis=1)#removing internal indexing
```

```python
travels_dataset.head(5)
```

```python
employee_expense = travels_dataset[['Employee','Actual Total Expenses']]
```

```python
employee_expense[employee_expense['Employee'].duplicated()]
```

```python
#groupby method on derived dataset employee_expense
employee_expense = employee_expense.groupby(by='Employee').sum()
```

```python
employee_expense
```

```python
employee_expense = employee_expense.drop(['index'],axis=1)#droping column index
```

```python
employee_expense = employee_expense.reset_index()
```

```python
employee_expense 
```

```python
employee_expense.merge(employee_df, how='inner',left_on='Employee', right_on='name')
```
